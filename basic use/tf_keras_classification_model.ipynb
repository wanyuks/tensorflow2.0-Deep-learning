```python
import matplotlib as mlt
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import sys
import os
import tensorflow as tf
from tensorflow import keras
import time
```

读取所用的数据

```python
fashion_mnist = keras.datasets.fashion_mnist
(x_train_all, y_train_all), (x_test, y_tset) = fashion_mnist.load_data()
x_val, x_train = x_train_all[:5000], x_train_all[5000:]
y_val, y_train = y_train_all[:5000], y_train_all[5000:]

print("x train shape:", x_train.shape)
print("y train shape:", y_train.shape)
print("x val shape:", x_val.shape)
print("x val shape:", x_val.shape)
print("x test shape:", x_test.shape)
print("x test shape:", x_test.shape)
```

了解数据

```python
def show_image(image):
    plt.imshow(image, cmap="binary")
    plt.show()
    
show_image(x_train[0])
```

```python
def show_images(n_rows, n_cols, x_data, y_data, class_names):
    assert len(x_data) == len(y_data)
    assert n_rows * n_cols < len(x_data)
    plt.figure(figsize=(n_rows * 2, n_cols * 3))
    for row in range(n_rows):
        for col in range(n_cols):
            index = row * col + col
            plt.subplot(n_rows, n_cols, index+1)
            plt.imshow(x_data[index], cmap="binary")
            plt.axis("off")
            plt.title(class_names[y_data[index]])
    plt.show()
class_names = ["T-shirt", "Trouser", "pullover", "Dress", "Coat",
              "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
show_images(4, 5, x_train, y_train, class_names)
```

使用tf.keras的Sequential构建模型

```python
"""# method1
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28, 28]))
model.add(keras.layers.Dense(300, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation='softmax'))"""

# method2
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation='softmax')
])


#sparse: 因为y只是一个index的值，需要将其one hot成一个10维的向量，所以用sparse
model.compile(loss="sparse_categorical_crossentropy",
             optimizer='adam',
             metrics=["accuracy"])
```

查看模型结构

```python
model.layers
```

```python
model.summary()
```

开始训练模型

```python
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_val, y_val))
```

查看history

```python
print("history type:",type(history))
print(history.history)
```

画出各指标的变化趋势

```python
def curves(history):
    pd.DataFrame(history.history).plot(figsize=(10, 6))
    plt.grid(True)
    plt.gca().set_ylim(0.1)
    plt.show()
curves(history)
```
